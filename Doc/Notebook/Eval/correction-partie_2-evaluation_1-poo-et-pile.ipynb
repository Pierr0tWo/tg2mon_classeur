```python
class Yaourt :
    
    def __init__(self,arome,duree) :
        
        # Assertions : question 1.a.
        assert arome in ["vanille", "fraise", "abricot", "aucun"], "Cet arome est inconnu"
        assert duree >= 1 and duree <=  365, "la durée doit être comprise entre 1 et  365"
        
        self.__arome = arome
        self.__duree = duree
        if arome == 'aucun' :
            self.__genre = 'nature'
        else :
            self.__genre = 'aromatise'
    
    # Méthode GetArome(self) question 1.c.
    def GetArome(self) :        
        return self.__arome
    
    def GetDuree(self):
        return(self.__duree)

    def GetGenre(self):
        return (self.__genre)
    
    # Mutateur de durée
    def SetDuree(self,duree) :
        if duree > 0 :
            self.__duree = duree
            
    def __SetGenre(self,arome):
        if arome == 'aucun':
            self.__genre = 'nature'
        else:
            self.__genre = 'aromatise'

    # Mutateur d’arôme SetArome(self,arome) - question 2.
    def SetArome(self, arome) :
        assert arome in ["aucun","vanille", "fraise", "abricot"], "Cet arome est inconnu"
        self.__arome = arome
        self.__SetGenre(arome)
```
